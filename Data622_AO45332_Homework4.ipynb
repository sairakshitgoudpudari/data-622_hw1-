```python
!pip uninstall -y transformers
!pip install -U spacy benepar 'transformers==4.1.1'
!python -m spacy download en_core_web_sm

import spacy
from spacy import displacy
import benepar
import nltk
from nltk import Tree

# Load models
nlp = spacy.load("en_core_web_sm")
benepar.download('benepar_en3')
nlp.add_pipe("benepar", config={"model": "benepar_en3"})

# The two famous opening sentences
sentence1 = "Four score and seven years ago our fathers brought forth on this continent, a new nation, conceived in Liberty, and dedicated to the proposition that all men are created equal."
sentence2 = "Now we are engaged in a great civil war, testing whether that nation, or any nation so conceived and so dedicated, can long endure."

print("=== 1. Tokenization (spaCy) ===\n")

doc1 = nlp(sentence1)
doc2 = nlp(sentence2)

print("Sentence 1 tokens:")
print([token.text for token in doc1])
print("\nSentence 2 tokens:")
print([token.text for token in doc2])

print("\n=== 2. Part-of-Speech Tagging (Sentence 1) ===\n")

print("Token".ljust(15), "POS".ljust(8), "Tag".ljust(6), "Explanation")
print("-" * 40)
for token in doc1:
    print(f"{token.text:<15} {token.pos_:<8} {token.tag_:<6} {spacy.explain(token.tag_) or ''}")

print("\n=== 3. Dependency Parsing (Sentence 2) ===\n")

print("Token".ljust(15), "Dep".ljust(12), "Head".ljust(15), "Head text")
print("-" * 55)
for token in doc2:
    head_text = token.head.text if token.head != token else "ROOT"
    print(f"{token.text:<15} {token.dep_:<12} {head_text:<15} {token.head.text}")

print("\n=== 4. Constituent Parsing (Sentence 1) using benepar ===\n")

# benepar adds ._.parse attribute to each sentence span
sent1_span = list(doc1.sents)[0]
parse_tree = sent1_span._.parse_string
print("Constituency parse tree (string format):")
print(parse_tree)

# Optional: pretty-print tree using NLTK Tree
tree = Tree.fromstring(parse_tree)
print("\nPretty-printed tree:")
tree.pretty_print()

print("\n=== 5. Noun Phrases (Noun Chunks) using spaCy ===\n")

print("Noun phrases in Sentence 1:")
for chunk in doc1.noun_chunks:
    print(f"• {chunk.text}  ({chunk.root.text}, {chunk.root.pos_})")

print("\nNoun phrases in Sentence 2:")
for chunk in doc2.noun_chunks:
    print(f"• {chunk.text}  ({chunk.root.text}, {chunk.root.pos_})")
```
