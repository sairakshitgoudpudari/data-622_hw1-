```python
import random
from typing import Dict, List

def build_markov_chain(words: List[str]) -> Dict[str, List[str]]:
    """
    Build a simple first-order Markov chain over words:
    current_word -> list of possible next words.
    """
    model: Dict[str, List[str]] = {}

    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]

        if current_word not in model:
            model[current_word] = []
        model[current_word].append(next_word)

    return model


def generate_story(
    model: Dict[str, List[str]],
    start_phrase: str,
    length: int = 200
) -> str:
    """
    Generate a story starting from 'start_phrase' using the given Markov model.
    """
    story_words = start_phrase.strip().split()

    current_word = story_words[-1]
    if current_word not in model:
        current_word = random.choice(list(model.keys()))
        story_words.append(current_word)

    for _ in range(length):
        possible_next = model.get(current_word)
        if not possible_next:
            current_word = random.choice(list(model.keys()))
            story_words.append(current_word)
            continue

        next_word = random.choice(possible_next)
        story_words.append(next_word)
        current_word = next_word

    return " ".join(story_words)

print("Exercise 1 functions defined.")
```

```python
# Sample fairy-tale style training text (you can expand or replace this)
fairy_tale_text = """
Once upon a time there was a kingdom filled with forests, rivers, and mountains.
In that kingdom lived a young princess who loved to explore the woods.
Every morning she walked along the river, talking to the birds and the trees.
One day, she found a mysterious path that she had never seen before.
Curious and brave, she followed the path deep into the forest.
There she discovered an old castle covered in vines and moss.
Inside the castle, she met a wise old wizard who had been waiting for a visitor.
The wizard told her that the kingdom would face a great storm.
Only someone with a kind heart and a curious mind could guide the people to safety.
The princess listened carefully and promised to help.
She returned to the kingdom and warned everyone about the coming storm.
Together, they strengthened their homes and shared food with one another.
When the storm finally arrived, the kingdom was ready.
Although the winds were strong and the rain was heavy, the people stayed safe.
After the storm, the kingdom grew greener and brighter than ever before.
The princess became known as a hero, not because she was perfect, but because she cared.
"""

# Tokenize by spaces for a simple Markov model
tokens = fairy_tale_text.split()

# Build the Markov chain
chain = build_markov_chain(tokens)

# Required starting phrase
start = "Once upon a time there was a kingdom..."

# Generate a Markov-chain story
markov_story = generate_story(chain, start_phrase=start, length=200)

print("=== Exercise 1: Markov Chain Story ===\n")
print(markov_story)
```

```python
from transformers import pipeline

def generate_llm_story() -> str:
    """
    Use a small Hugging Face model to create a story
    answering 'Why study data science today?'.
    """
    # Use a small GPT-2 style model that fits in Colab free tier
    generator = pipeline(
        "text-generation",
        model="distilgpt2"
    )

    prompt = (
        "Write a short, imaginative story that answers the question "
        "'Why study data science today?'. The story should have a main "
        "character, a simple plot, and a hopeful ending.\n\nStory:\n"
    )

    outputs = generator(
        prompt,
        max_length=350,
        num_return_sequences=1,
        do_sample=True,
        top_p=0.9,
        temperature=0.9
    )

    story = outputs[0]["generated_text"]
    return story

print("Exercise 2 function defined.")
```

```python
print("=== Exercise 2: LLM-Generated Story ===\n")
llm_story = generate_llm_story()
print(llm_story)
```
