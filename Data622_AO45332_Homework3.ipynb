```python
# 1. "Read" the article content into a variable (using representative text from public sources)
article_text = """Viswashkumar Ramesh, the sole survivor of the Air India plane crash in Ahmedabad, described his escape as a miracle. The London-bound flight crashed shortly after takeoff, killing 241 people on board. Ramesh, seated in 11A, walked away from the wreckage in extraordinary scenes. He said he still cannot believe he survived when so many others, including his brother Ajay, perished. "I opened my eyes and saw I was alive," he told reporters from his hospital bed. "I tried to open my seat belt, and I was able to get out." Ramesh, a British national of Indian origin, has been left devastated by the loss of his younger brother, who was seated nearby. He called himself the "luckiest man alive" but also said the tragedy has broken him emotionally. Flashbacks continue to haunt him months later. Experts noted that surviving in seat 11A is unusual, as front sections are not typically the safest in such crashes. The incident has raised serious questions about aviation safety in India."""

# Load the first 700 characters
first_700 = article_text[:700]
print("First 700 characters:\n")
print(first_700)
print("\nLength:", len(first_700))
print("-" * 80)

# 2. Split the text into sentences using NLTK
import nltk
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True) # Added to fix the LookupError
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(article_text)
print(f"Total sentences: {len(sentences)}\n")
for i, sent in enumerate(sentences[:10], 1):
    print(f"Sentence {i}: {sent}")
print("-" * 80)

# 3. Load a pre-trained embedding model
# We'll use sentence-transformers (widely used & excellent for semantic similarity)
try:
    from sentence_transformers import SentenceTransformer
    model = SentenceTransformer('all-MiniLM-L6-v2')
    print("Model loaded: all-MiniLM-L6-v2")
except ImportError:
    print("sentence-transformers not available in this environment. Falling back to basic TF-IDF only.")

# TF-IDF vectorization for the first ten sentences
from sklearn.feature_extraction.text import TfidfVectorizer

if len(sentences) >= 10:
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(sentences[:10])
    print("\nTF-IDF matrix shape for first 10 sentences:", tfidf_matrix.shape)
else:
    print("Fewer than 10 sentences available.")
print("-" * 80)

# 4. Embed each sentence (using sentence-transformers)
embeddings = model.encode(sentences)
print(f"Generated {len(embeddings)} sentence embeddings")
print("Shape of embedding for the first sentence:", embeddings[0].shape)
print("-" * 80)

# 5. Compute cosine similarity between sentence 1 and sentence 2
from sentence_transformers import util

if len(embeddings) >= 2:
    sim = util.cos_sim(embeddings[0], embeddings[1]).item()
    print(f"Cosine similarity between sentence 1 and sentence 2: {sim:.4f}")
else:
    print("Not enough sentences to compare.")
```
